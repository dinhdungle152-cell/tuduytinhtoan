Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là 
độ dài ba cạnh hay không, nếu  không thì in  ra màn  hình  'Khong la ba  canh 
cua mot tam  giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, 
Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn 
hình.

```python
import math
a = float(input())
b = float(input())
c = float(input())
chu_vi = a + b + c
p = chu_vi / 2
dien_tich = math.sqrt(p * (p - a) * (p - b) * (p - c))
print(chu_vi)
print(dien_tich)
if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print("Tam giac deu")
    elif a == b or b == c or a == c:
        print("Tam giac can")
    elif a ** 2 + b ** 2 == c ** 2 or a ** 2 + c ** 2 == b ** 2 or b ** 2 + c ** 2 == a ** 2:
        print("Tam giac vuong")
    else:
        print("Tam giac thuong")
else:
    print("Khong la ba canh cua mot tam giac")
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
import math
a = float(input())
b = float(input())
c = float(input())
if a == 0:
    if b == 0:
        if c == 0:
            print("Phương trình vô số nghiệm")
        else:
            print("Phương trình vô nghiệm")
    else:
        x = -c / b
        print("Nghiệm:", x)
else:
    delta = b ** 2 - 4 * a * c
    if delta > 0:
        x1 = (-b + math.sqrt(delta)) / (2 * a)
        x2 = (-b - math.sqrt(delta)) / (2 * a)
        print("Hai nghiệm phân biệt:", x1, "và", x2)
    elif delta == 0:
        x = -b / (2 * a)
        print("Nghiệm kép:", x)
    else:
        print("Phương trình vô nghiệm")
```

Bài 8: Giải hệ phương trình tuyến tính: 
ax + by = m 
cx + dy = n 

```python
a = float(input())
b = float(input())
c = float(input())
d = float(input())
m = float(input())
n = float(input())
D = a*d - b*c
Dx = m*d - b*n
Dy = a*n - m*c
if D != 0:
    x = Dx / D
    y = Dy / D
    print("Nghiệm: x =", x)
    print("Nghiệm: y =", y)
else:
    if Dx == 0 and Dy == 0:
        print("Hệ phương trình vô số nghiệm")
    else:
        print("Hệ phương trình vô nghiệm")
```

Bài 9: Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết 
ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao 
nhiêu giây. 

```python
x = int(input())
gio = x // 3600
phut = (x % 3600) // 60
giay = (x % 3600) % 60
print(gio, "giờ", phut, "phút", giay, "giây")
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một 
điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
x0 = float(input())
y0 = float(input())
R = float(input())
x = float(input())
y = float(input())
left = (x - x0)**2 + (y - y0)**2
right = R**2
if left == right:
    print("Điểm A thuộc đường tròn")
elif left < right:
    print("Điểm A nằm trong đường tròn")
else:
    print("Điểm A nằm ngoài đường tròn")
```
