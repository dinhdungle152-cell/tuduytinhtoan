Bài 1: Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diện tích và chu 
vi của nó.

```python
chieu_dai = float(input())
chieu_rong = float(input())
a = chieu_dai
b = chieu_rong
chu_vi = (a + b) * 2
dien_tich = a * b
print("chu vi hình chữ nhật là:", chu_vi)
print("diện tích hình chữ nhật là:", dien_tich)
```
