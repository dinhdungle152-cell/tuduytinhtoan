Bài 2b: Tìm số nhỏ nhất của 1 list và vị trí của nó, nếu có nhiều số nhỏ nhất bằng nhau thì lấy chỉ số đầu tiên (nhỏ nhất)
10 0 0 12 34 5 6 7  8 89 0 1 0 1

```python
def min_and_pos(lst):
    if not lst:
        return None
    val = lst[0]
    pos = 0
    for i, x in enumerate(lst):
        if x < val:
            pos = 1
            val = x
    return val, pos + 1

#assert min_and_pos([10, 0, 0, 12, 34, 5, 6, 7, 8, 89, 0, 1]) == (0, 2)
#assert min_and_pos([]) == None

print("Pass all Tests.")
```

Bài 3: Tìm giá trị trung bình của các số nguyên dương trong một list.
Nếu không có số dương nào thì trả về "invalid"

```python
def Average_positive_numbers_of_list(lst):
    s = 0
    count = 0
    for x in lst:
        if x > 0:
            count += 1
            s += x
    if count == 0:
        return "invalid"
    return round(s/count, 2)
    
#assert Average_positive_numbers_of_list([1, 2, 6, 1, -10, -10]) == 2.50
#assert Average_positive_numbers_of_list([-2, -1, -20, -343]) == "invalid"
#assert Average_positive_numbers_of_list([1]) == "invalid"

print("Accepted!!!")
```

Bài 4: Kiểm tra một list số nguyên có phải là đối xứng hay không
[1, 2, 3, 5, 3, 2, 1]

```python
def kt_doi_xung(lst):
    if not lst:
        return None
    for i in range(len(lst) // 2):
        if lst[i] != lst[-i - 1]:
            return False
    return True
 
assert kt_doi_xung([1, 2, 3, 5, 3, 2, 1]) == True
assert kt_doi_xung([1, 2, 3, 5, 3, 4, 1]) != True
assert kt_doi_xung([]) == None

print("Accepted!!")
```

Bài 5: Đưa tất cả các số bằng 0 trong list về đầu của list. Các số khác không thay đổi thứ tự trước sau
3 4 6 7 1 9 0 0 23 0 0 34 67 89 0 0 0 0 3 4 6 7 1 9 23 34 67 89

```python
def move_zero_to_first(lst):
    zero_lst = []
    None_zero_lst = []
    for x in lst:
        if x:
            None_zero_lst.append(x)
        else:
            zero_lst.append(x)
    zero_lst.extend(None_zero_lst)
    return zero_lst

assert move_zero_to_first([3, 4, 0, 0, 23, 0, 0, 34, 67, 89]) == [0, 0, 0, 0, 3, 4, 23, 34, 67, 89]
assert move_zero_to_first([0, 0, 0, 0]) == [0, 0, 0, 0]
assert move_zero_to_first([3, 4, 0, 0, 1, 2, 3]) == [0,0,3,4,1,2,3]

print("accepted!!!")
```

Bài 6: Xóa tất cả các số giống nhau trong list, giữ lại đúng 1 lần, Các số khác giữ nguyên vị trí trước 
[1, 2, 3, 4, 1, 2, 3, 4, 1, 1, 4, 3] 

```python
def unique_array(lst):
    if not lst:
        return None
    unique_lst = []
    for x in lst:
        if x not in unique_lst:
            unique_lst.append(x)
    return unique_lst

assert unique_array([1, 2, 3, 4, 1, 2, 3, 4, 1, 1, 4, 3]) == [1,2,3,4]
assert unique_array([1, 1, 1, 1]) == [1]
assert unique_array([]) == None

print("Accepted!!!")
```

Bài 7: Kiểm tra một list số nguyên có phải là dãy tăng, dãy giảm hay vô danh
1 3 5 6 7 8 9: Tang
9 8 7 6 5 4 0: Giam
1 2 3 7 1 0 2: Vo danh
1 2 2 4 5 7: Vo danh

```python
def check_order_of_list(lst):
    if not lst:
        return None
    giam = 1
    tang = 1
    for i in range(len(lst) - 1):
        if lst[i] > lst[i + 1]:
            giam += 1
        elif lst[i] < lst[i + 1]:
            tang += 1

    if giam == len(lst):
        return "Giam"
    if tang == len(lst):
        return "Tang"
    
    return "Vo danh"

assert check_order_of_list([1, 3, 5, 6, 7, 8, 9]) == "Tang"
assert check_order_of_list([9, 8, 7, 6, 5, 4, 0]) == "Giam"
assert check_order_of_list([1, 2, 3, 7, 1]) == "Vo danh"
assert check_order_of_list([7, 6, 5, 4, 3, 3]) == "Vo danh"
print("Pass all tests!!!")
```

Bài 8: Kiểm tra một danh sách số nguyên có phải là ziczac khong
1 3 2 5 2 8 7 9 2 100: ziczac
34 2 56 7 90 20 30 7 120: ziczac
2 2 34 4 56 7 89 : không phải ziczac

```python
def is_ziczac(lst):
    n = len(lst)
    if n < 2:
        return True
    if lst[0] == lst[1]:
        return False
    for i in range(1, n - 1):
        if (lst[i] - lst[i-1]) * (lst[i] - lst[i+1]) <= 0:
            return False
    return True

assert is_ziczac([1, 3, 2, 5, 0, 7, 6, 8,0]) == True
assert is_ziczac([34, 2, 56, 7, 80, 12, 24, 14]) != False
assert is_ziczac([2, 2, 34, 4, 56, 7, 89]) == False

print("All accepted!")
```

Bài 9: Tìm và in ra tất cả các vị trí của số nguyên k trong một danh sách số nguyên cho trước. Nếu không có k trong ds thì in ra "-1".
VD: k = 3, [1,3,4,4,3,3,4,3,6,1] => k xuất hiện tại các vị trí (trả về 1 list) [1, 4, 5, 7]
k = 10, [1,3,4,4,3,3,4,3,6,1] => k không xuất hiện trong ds => trả về None

```python
def find_pos_of_k(lst, k):
    if not lst:
        return None
    pos_lst = []
    for i, x in enumerate(lst):
        if x == k:
            pos_lst.append(i)
    return pos_lst if pos_lst else None

assert find_pos_of_k([1,3,4,4,3,3,4,3,6,1], 3) == [1, 4, 5, 7]
assert find_pos_of_k([1,3,4,4,3,3,4,3,6,1], 10) == None
assert find_pos_of_k([1, 1, 2, 3], 3) == [3]

print("Tests accepted!")
```

Bài 11: Cho một danh sách các số nguyên có giá trị trong đoạn [0,9]. Liệt kê số lần xuất hiện của các số có mặt trong danh sách. Số nào xuất hiện trước thì liệt kê trước
VD 1 2 1 2 1 1 1 1
output: 1: 6 2: 2

```python
def list_number_in_lst(lst):
    if not lst:
        return None
    count_dict = {}
    for x in lst:
        if x in count_dict:
            count_dict[x] += 1
        else:
            count_dict[x] = 1
    return count_dict

assert list_number_in_lst([1, 2, 1, 1, 2, 1, 1]) == {1: 5, 2: 2}
assert list_number_in_lst([1, 2, 1, 1, 2, 1, 1, 3, 3, 4, 2, 1, 2, 3]) == {1: 6, 2: 4, 3: 3, 4: 1}
assert list_number_in_lst([]) == None
assert list_number_in_lst([4, 3, 1, 2, 1, 1, 2, 1, 1])== {4: 1, 3: 1, 1: 5, 2: 2}

print("Pass all tests!")
```

Bài 11B: Cho một danh sách các số nguyên có giá trị trong đoạn [0,9]. Liệt kê số lần xuất hiện của các số có mặt trong danh sách. Số nào nhỏ thì liệt kê trước
VD 2 1 2 1 2 1 1 1 1
output: dict 

```python
def list_number_in_lst_order(lst):
    if not lst:
        return None
    lst.sort()
    count_dict = {}
    for x in lst:
        count_dict[x] = count_dict.get(x, 0) + 1
    return count_dict
    
assert list_number_in_lst_order([2, 3, 3, 1, 1, 2, 1, ]) == {2: 2, 3: 2, 1: 3}
assert list_number_in_lst_order([2, 3, 8, 8, 8, 3, 1, 1, 2, 1, 4, 3, 4, 5]) == {2: 2, 3: 3, 8: 3, 1: 3, 4: 2, 5: 1}

print("Dung roi")
```

Bài 12: Cho một chuỗi kí tự chỉ gồm các kí tự chữ cái thường 'a'....'z'. Hãy liệt kê số lần xuất hiện của các kí tự có mặt trong chuỗi theo thứ tự tăng dần.

output:
a : 2
