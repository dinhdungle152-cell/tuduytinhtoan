Bài 12: Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k) 
lớn nhất nhưng nhỏ hơn n

```python
n = int(input())
k = 0
s = 0
while s + (k + 1) < n:
    k += 1
    s += k
print("k =", k)
print("S(k) =", s)
```

Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho 
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
A = int(input())
S = 0
n = 0
while S <= A:
    n += 1
    S += 1 / n
print("n nhỏ nhất là:", n)
```

Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập 
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng

```python
max_val = None
min_val = None
while True:
    x = int(input())
    if x == -1:
        break
    if max_val is None:  
        max_val = min_val = x
    else:
        if x > max_val:
            max_val = x
        if x < min_val:
            min_val = x
if max_val is None:
    print("Không có dữ liệu!")
else:
    print("Số lớn nhất:", max_val)
    print("Số nhỏ nhất:", min_val)
```

Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu 
chữ số lẻ

```python
n = int(input())
chan = 0
le = 0
while n > 0:
    digit = n % 10   
    if digit % 2 == 0:
        chan += 1
    else:
        le += 1
    n //= 10  
print("Số chữ số chẵn:", chan)
print("Số chữ số lẻ:", le)
```

Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay 
không 

```python
n = int(input())
if n <= 0:
    print("Không phải dạng 3^k")
else:
    while n % 3 == 0:
        n //= 3
    if n == 1:
        print("Là dạng 3^k")
    else:
        print("Không phải dạng 3^k")
```

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số 
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong 
dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
A = int(input())
f1, f2 = 1, 1
while f2 <= A:
    f1, f2 = f2, f1 + f2
print("Số Fibonacci lớn nhất <= A là:", f1)
```

Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố.

```python
n = int(input())
dem = 0
while n > 0:
    digit = n % 10
    if digit in [2, 3, 5, 7]:
        dem += 1
    n //= 10
print("Số chữ số nguyên tố:", dem)
```
