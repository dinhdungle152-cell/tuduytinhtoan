W3A1 Nhập vào một số nguyên dương có 3 chữ số. Tìm và in ra màn hình “SỐ 
NGUYÊN” đảo ngược của n. Không dùng kiểu chuỗi. Chú ý: kiểm tra chỉ xử 
lý các số nguyên dương có đúng 3 chữ số. Các số khác không xử lý.

```python
n = int(input())
if 100 <= n <= 999:
    tram = n // 100
    chuc = (n // 10) % 10
    don_vi = n % 10
    so_dao_nguoc = don_vi * 100 + chuc * 10 + tram
    print(so_dao_nguoc)
else:
    print("không phải số có 3 chữ số")
```

W3A3 Kiểm tra xem một số có phải là lũy thừa của 2 hay không 
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 
hay không. 
Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0

```python
n = int(input("Nhập số n: "))
if n > 0 and (n & (n - 1)) == 0:
    print(n, "là lũy thừa của 2")
else:
    print(n, "không phải là lũy thừa của 2")
```

W3A11 [Phân loại tam giác] 
Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm 
tra 3 số đã nhập có tạo thành một tam giác hợp lệ hay không. Nếu không in 
ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", 
"Tam giác cân", hay "Tam giác thường". 

```python
a = int(input())
b = int(input())
c = int(input())
if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print("Tam giac deu")
    elif a == b or b == c or a == c:
        print("Tam giac can")
    elif a ** 2 + b ** 2 == c ** 2 or a ** 2 + c ** 2 == b ** 2 or b ** 2 + c ** 2 == a ** 2:
        print("Tam giac vuong")
    else:
        print("Tam giac thuong")
else:
    print("Khong la ba canh cua mot tam giac")
```

W3A14 [Giải phương trình bậc nhất] 
Nhập vào hai số thực a và b của phương trình ax + b = 0. Tìm và in ra 
nghiệm của phương trình, kết quả làm tròn đến số thập phân thứ 2. In ra "Vô 
nghiệm" hoặc "Vô số nghiệm" trong trường hợp tương ứng. 


```python
a = float(input())
b = float(input())
if a != 0:
    x = -b / a
    print(f"Nghiệm của phương trình là: {x:.2f}")
else:
    if b == 0:
        print("Vô số nghiệm")
    else:
        print("Vô nghiệm")
```

W3A15 [Xếp loại học lực] 
Nhập vào điểm trung bình của một học sinh và in ra học lực của học sinh đó. 
Xếp loại học lực theo quy tắc: - >= 8.0: Giỏi - >= 6.5: Khá - >= 5.0: Trung bình - < 5.0: Yếu

```python
diem = float(input())
if diem >= 8:
    print("Giỏi")
elif 6.5 <= diem < 8:
    print("Khá")
elif 5 <= diem < 6.5:
    print("Trung bình")
else:
    print("Yếu")
```
