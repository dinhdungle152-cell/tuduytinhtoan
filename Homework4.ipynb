W4A20 Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không 

```python
n = int(input())
if n <= 0:
    print("Không phải dạng 2^k")
else:
    while n % 2 == 0:
        n //= 2
    if n == 1:
        print("Là dạng 2^k")
    else:
        print("Không phải dạng 2^k")
```

W4A21 Nhập vào số nguyên dương n, tính tổng các chữ số của n 

```python
n = int(input())
tong = 0
while n > 0:
    tong += n % 10
    n //= 10
print("tổng các chữ số của n là:", tong)
```

W4A22 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input())
chan = 0
le = 0
while n > 0:
    digit = n % 10   
    if digit % 2 == 0:
        chan += 1
    else:
        le += 1
    n //= 10  
print("Số chữ số chẵn:", chan)
print("Số chữ số lẻ:", le)
```

W4A25 Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. 
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python
max_val = None
min_val = None
while True:
    x = int(input())
    if x == -1:
        break
    if max_val is None:  
        max_val = min_val = x
    else:
        if x > max_val:
            max_val = x
        if x < min_val:
            min_val = x
if max_val is None:
    print("Không có dữ liệu!")
else:
    print("Số lớn nhất:", max_val)
    print("Số nhỏ nhất:", min_val)
```

W4A26 [Fibonaci1] 
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ 
bằng tổng hai số trước đó 
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quas A

```python
A = int(input())
f1, f2 = 1, 1
while f2 <= A:
    f1, f2 = f2, f1 + f2
print("Số Fibonacci lớn nhất <= A là:", f1)
```

W4A27 Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có 
ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng) 
Dùng cách: for trên index; KHÔNG dùng hàm split(), len() 

```python
s = input()
dem = 0
i = 0
for c in s:   
    if c != ' ':
        if i == 0 or s[i - 1] == ' ':
            dem += 1
    i += 1
print("Số từ trong chuỗi là:", dem)
```

W4A29 Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành 
thường và thường thành hoa. 

```python
s = input()
ket_qua = ""
for i in range(len(s)):
    if 'a' <= s[i] <= 'z':
        ket_qua += chr(ord(s[i]) - 32)
    elif 'A' <= s[i] <= 'Z':
        ket_qua += chr(ord(s[i]) + 32)
    else:
        ket_qua += s[i]
print("Chuỗi sau khi đổi:", ket_qua)
```

W4A30 Nhập vào một chuỗi có dạng 3 số nguyên, mỗi số nguyên cách nhau một dấu phẩy, hãy tính 
tổng 3 số nguyên đó 
VD: 
Nhập: 3, 12, 15 
Tổng: 

```python
s = input()
tong = 0
so = ""
for i in range(len(s)):
    if s[i] != ',':
        if s[i] != ' ':
            so += s[i]
    else:
        tong += int(so)
        so = ""
if so != "":
    tong += int(so)
print("Tổng:", tong)
```

W4A31 Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in 
thường, bao nhiêu ký tự số

```python
s = input()
dem_hoa = 0
dem_thuong = 0
for c in s:
    if c.isupper():
        dem_hoa += 1
    elif c.islower():
        dem_thuong += 1
print("In hoa:", dem_hoa)
print("In thường:", dem_thuong)
```
